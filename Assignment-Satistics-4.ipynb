```python
print('Problem Statement 3: Calculate F Test for given 10, 20, 30, 40, 50 and 5,10,15, 20, 25 ')
print('')
import numpy as np
x1=[10, 20, 30, 40, 50]
x2=[5,10,15, 20, 25]
print('mean x1 bar for first set is',np.mean(x1))
print('mean x2 bar for second set is',np.mean(x2))
print('')
print('variance x1 square for first set is',np.var(x1))
print('variance x2 square for second set is',np.var(x2))
print('')
print('formula for F variance is F=S1/S2 where S1 is the largest variance and S2 is the smalles')
print('')
print('F test value is ',np.var(x1)/np.var(x2))
print('degree of freedom of numerator is (no of sample -1)')
print('                                  (5 -1)')
print('                                  4')
print('degree of freedom of denominator is (no of sample -1)')
print('                                  (5 -1)')
print('                                  4')
print('F test table value is 6.3882')
print('')
print('Since F test value 6.3882 is greater than 4 , We dont have enough evidence to reject null hypothesis')
```
